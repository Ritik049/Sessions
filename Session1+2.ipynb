# Combining and Merging

# Data Manipulation Techniques in Pandas

We will explore key data manipulation techniques using the Pandas library in Python: combining, merging, grouping, and aggregating. These techniques are essential for effective data analysis.

## 1. Combining DataFrames

Combining DataFrames refers to joining multiple DataFrames into one. This can be done using concatenation.

### Concatenation
Concatenation allows you to stack DataFrames either vertically (adding rows) or horizontally (adding columns).



````python
# Example DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenating vertically (stacking rows)
combined_vertical = pd.concat([df1, df2], axis=0)
print("Combined DataFrame (Vertical):")
print(combined_vertical)
````

## 2. Merging DataFrames
Merging is a more complex operation that combines DataFrames based on a common key or index, similar to SQL joins. It allows for inner, outer, left, and right joins.

````python
# Example DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value2': [4, 5, 6]})

# Merging on the key
merged_inner = pd.merge(df1, df2, on='key', how='inner')  # Inner join
print("\nMerged DataFrame (Inner Join):")
print(merged_inner)
````

## 3. Grouping DataFrames
Grouping involves segmenting the DataFrame into subsets based on the values of one or more columns. This is often done before applying an aggregation function.

````python
# Example DataFrame
df = pd.DataFrame({
    'category': ['A', 'A', 'B', 'B'],
    'value': [1, 2, 3, 4]
})

# Grouping by category
grouped = df.groupby('category')
print("\nGrouped DataFrame:")
print(grouped)
````

## 4. Aggregating Data
Aggregating refers to applying a function (like sum, mean, count, etc.) to each group created by the grouping operation. It summarizes the data.



````python
# Aggregating the grouped data
aggregated_sum = grouped['value'].sum()  # Sum of values for each category
print("\nAggregated DataFrame (Sum):")
print(aggregated_sum)
````

## Summary of Terms
- Combining: Stacking DataFrames together (vertically or horizontally).
- Merging: Joining DataFrames based on common keys, similar to SQL joins.
- Grouping: Segmenting data into groups based on specified criteria.
- Aggregating: Applying summary statistics to groups of data.
